1. Write a function which takes an input parameter $A$, $i$ and $j$ and returns the dot product of the $i$th and $j$th row (indexing starts at 0).
             

```python
import numpy as np

def dot_prod(A, i, j):
    '''Dot product of i th & j th row of A'''
    return np.dot(A[i, :], A[j, :])




A= np.array( [[1, 2, 3],
              [1, 1, 1],
              [3, 4, 5]] )

print(dot_prod(A, 0, 0))
print(dot_prod(A, 0, 1))
print(dot_prod(A, 0, 2))
```

2. Compute the matrix equation $AB + 2B^2 - I$ for matrices $A = \begin{bmatrix} 3 & 4 \\\ -1 & 2 \end{bmatrix}$ and $B = \begin{bmatrix} 5 & 2 \\\ 8 & -3 \end{bmatrix}$. 

```python
A = np.array([[3, 4],
              [-1, 2]])
B = np.array([[5, 2],
              [8, -3]])
I = np.eye(2)

AB = A@B
B2 = B@B

print(AB + 2*B2 - I)
```

### 2. Given that $A=\begin{bmatrix} 1 & 2 \\\ 3 & 0  \end{bmatrix} $, and $b=\begin{bmatrix} 4 \\\ 6\end{bmatrix} $.


#### (a) Find $A^{-1}$.


Let $B = \begin{bmatrix} a & b \\\ c & d  \end{bmatrix} $, we have $AB=I$, which gives 


$\begin{bmatrix} 1 & 2 \\\ 3 & 0  \end{bmatrix} \begin{bmatrix} a & b \\\ c & d  \end{bmatrix}  = \begin{bmatrix} 1 & 0 \\\ 0 & 1  \end{bmatrix}$.

We can expand the matrix formulation to get the system of equations as 


$a+2c = 1, \\ b+2d=0, \\3a = 0, \\3b=1$.

By solving this equation, we have

$A^{-1} = B = \begin{bmatrix} 0 & \frac{1}{3} \\\ \frac{1}{2} & -\frac{1}{6} \end{bmatrix} $.

We can also do this using Linear Algebra package in SciPy,

```python
import scipy.linalg as la


A=np.array([[1, 2],
            [3, 0]])
b=np.array([4, 6]).reshape(2,1)


Ainv = la.inv(A)
print(Ainv)

print(A@Ainv)
```

#### (b) Use $A^{-1}$ to solve $Ax=b$ for x. Check your answer by computing Ax.

Multiply both side of $Ax=b$ with $A^{-1}$, we have 

$A^{-1}Ax=A^{-1}b$, 

which gives $x=A^{-1}b$.

```python
x= Ainv@b
print(x)

Ax = A@x
print(Ax)
```

### 3. Given the following matrices:
$A=\begin{bmatrix} 2 &0 \\\ 0 &2 \end{bmatrix} $, $B=\begin{bmatrix} -2 &0 \\\ 0 &2 \end{bmatrix}$, $C=\begin{bmatrix} 0 &0 \\\ 0 & 1 \end{bmatrix}$,  $D=\begin{bmatrix} 1 & 2 \\\ 0 & 1 \end{bmatrix}$.

Draw and label each of the following:

$Ax, Bx, Cx, Dx$


```python
A=np.array([[2, 0],
            [0, 2]])
B=np.array([[-2, 0],
            [0, 2]])
C=np.array([[0, 0],
            [0, 1]])
D=np.array([[1, 2],
            [0, 1]])

x=np.array([2,2]).reshape(2,1)

Ax=A@x
Bx=B@x
Cx=C@x
Dx=D@x

print(Ax)
```

```python
import matplotlib.pyplot as plt

origin = np.array([[0], [0]])

plt.style.use('default')
dpi=300
figsize=(8, 5)
xmin=-7;
xmax=8;
ymin=-2;
ymax=5;

plt.clf()
plt.figure(figsize=figsize, dpi=dpi)

plt.quiver(origin[0,0], origin[1,0], x[0,0], x[1,0], scale_units='xy', angles='xy', scale=1, label='$x$')
plt.quiver(origin[0,0], origin[1,0], Ax[0,0], Ax[1,0], scale_units='xy', angles='xy', scale=1, label='$Ax$', color='r')
plt.quiver(origin[0,0], origin[1,0], Bx[0,0], Bx[1,0], scale_units='xy', angles='xy', scale=1, label='$Bx$', color='b')
plt.quiver(origin[0,0], origin[1,0], Cx[0,0], Cx[1,0], scale_units='xy', angles='xy', scale=1, label='$Cx$', color='y')
plt.quiver(origin[0,0], origin[1,0], Dx[0,0], Dx[1,0], scale_units='xy', angles='xy', scale=1, label='$Dx$', color='c')

plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)

plt.legend()
plt.grid()

plt.show()
```
